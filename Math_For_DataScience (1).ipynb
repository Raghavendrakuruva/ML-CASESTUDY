```python
import numpy as np

left_points = [(1,1), (2,1)]
right_points = [(5,4), (4,3)]
on_line = [(3,3), (2,4)]
q = (2,4)

col = [left_points, right_points, on_line, q]

n = np.array([1, 1])  # normal vector
l = []                # to store results

for item in col:
    # item can be a list of points OR a single point
    if isinstance(item, tuple):      # single point like (2,4)
        points = [item]
    else:                            # list of points like [(1,1),(2,1)]
        points = item

    for x, y in points:
        p = np.array([x, y])
        d = np.dot(n, p) - 6   # x + y - 6

        if d > 0:
            side = 'rightside'
        elif d < 0:
            side = 'leftside'
        else:
            side = 'on_line'

        l.append([(x, y), d, side])

print(l)
```

```python
import pandas as pd
```

```python
pd.DataFrame(l,columns=['point','value','classification'])
```

### Task 5: Classify Points and Compute Distances for Hyperplanes

#### 1: Calculate the euclidean distance between these 2 points {p1=(10,1,2,3), p2=(11,10,2,5)}

#### 2: 4D Hyperplane
- Hyperplane: 2x1 + x2 - x3 + x4 - 5 = 0
- Points: P1(1,1,0,2), P2(2,0,1,1), P3(1,2,1,1), q(1,1,1,1)
- Compute:
  1. Dot product: n · p + b , where n is the normal vector and p is the data point b is the intercept.
  2. Euclidean distance from hyperplane
  3. Side (Positive/Negative/On)
- Document results in a table:

| Point | Value | Classification |
|-------|-------|----------------|


```python
a=(1,1,0,2)
b=(2,0,1,1)
c=(1,2,1,1)
q=(1,1,1,1)

col = [a, b,c, q]

n = np.array([2,1,-1,1])  # normal vector
l = []                # to store results

for item in col:
        p = np.array(item)
        print(p)
        d = np.dot(n, p) - 6   # x + y - 6

        if d > 0:
            side = 'rightside'
        elif d < 0:
         side = 'leftside'
        else:
            side = 'on_line'

        l.append([p, d, side])

print(l)
```
