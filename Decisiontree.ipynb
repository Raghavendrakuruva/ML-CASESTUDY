# Part A: Conceptual Questions


## 1.What is a Decision Tree and how does it make decisions?

```python
Decision tree is non_parametric algo and supervised learning,it divides the data into multiple 
chunks based on range of feature value

Decision Tree make decision like same as human thinking that is like if else sequences of asking 
questions 
```

## 2.What does a split criterion mean in a Decision Tree?

```python
In a Decision Tree, a split criterion is the rule used to decide how a node is split into child nodes.
It measures how good a feature and a threshold are at separating the data into more “pure” subsets.

The split criterion tells the tree where and how to split the data at each node.
```

## 3.Difference between Gini Impurity and Entropy.

```python
Gini Impurity-it is used for split criterion for classifiaction problem it based on features
              maximum information gain decision tree split,it computationally easy,it consider
              all classification errors while splitting
               
entropy-it is used for split criterion for classifiaction problem it based on features
              maximum information gain decision tree split,it computationally expensive,
               due it deals with  logarthims  it consider
              all classification errors while splitting
```

## 4.Why do Decision Trees tend to overfit?

```python
Decision Tree is follow Greedy approach it means keepon asking questions and splitting the data
to pure nodes or furthur split is not possible so due to this it memorizes the whole by in the
splitting process this is the reason decision tree tend to overfit
```

## 5.What role does the max_depth parameter play?

```python
Decision tree tends to overfit so to address this splitting can stop with in minimum steps
by using hyper-parameter like max_depth,computation also becomes easy by reducing the steps
low max_depth:
-------------
High bias, low variance

May underfit the data

large max_depth:
----------------
Low bias, high variance

May overfit the data
```

```python
from sklearn.model_selection import train_test_split
```

# Part D: Model Building

```python
Create a Decision Tree Classifier with:
criterion='gini'
random_state=42
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt_model = DecisionTreeClassifier(
    criterion='gini',
    random_state=42
)
dt_model
```

```python
Train the model on training data
```

```python
dt_model.fit(x_train,y_train)
```

```python
## Predict on the test data.
```

```python
y_predict=dt_model.predict(x_test)
y_predict
```

#  Part E: Model Evaluation
