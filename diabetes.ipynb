```python
data['Outcome']=data['Outcome'].astype('category')# outcome change to category dtype
```

```python
data.info()
```

```python
## most of columns contain outliers so we are impute nan with median
```

```python
l=list(data.columns)[:8]
```

```python
l
```

```python
for i in l:
    data[i]=data[i].fillna(data[i].median())
```

```python
data.isna().sum()
```

```python
## outcome is category so we are replacing with mode
data['Outcome']=data['Outcome'].fillna(data['Outcome'].mode()[0])
```
