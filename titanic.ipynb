```python
## memory optimize
```

```python
titanic = titanic.astype({
    'Survived': 'category',
    'Pclass': 'category',
    'Sex': 'category',
    'Age':'int32',
    'SibSp': 'category',
    'Parch': 'category',
    'Cabin': 'category',
    'Embarked': 'category'
})                        
```

```python
titanic.info()## memory usage reduce from 88kb to 44kb
```

```python
## finding duplicates 
```

```python
titanic.duplicated().sum() ## there are no duplicates present in the dataset
```

```python
## visulaize
```

```python
# Set up the visualization style
plt.style.use('seaborn-v0_8')
fig, axes = plt.subplots(2, 2, figsize=(15, 12))

# Plot 1: Survival by Gender
gender_survival = titanic.groupby('Sex')['Survived'].size()
axes[0,0].bar(gender_survival.index, gender_survival.values , color=['lightblue', 'pink'])
axes[0,0].set_title('Survival Rate by Gender', fontsize=14, fontweight='bold')
axes[0,0].set_ylabel('Survival Rate (%)')

# Plot 2: Survival by Passenger Class
class_survival = titanic.groupby('Pclass')['Survived'].size()
axes[0,1].bar(class_survival.index, class_survival.values * 100, color=['gold', 'silver', 'brown'])
axes[0,1].set_title('Survival Rate by Passenger Class', fontsize=14, fontweight='bold')
axes[0,1].set_ylabel('Survival Rate (%)')

# Plot 3: Age distribution of survivors vs non-survivors
survived_age = titanic[titanic['Survived'] == 1]['Age'].dropna()
not_survived_age = titanic[titanic['Survived'] == 0]['Age'].dropna()
axes[1,0].hist([survived_age, not_survived_age], bins=20, label=['Survived', 'Did not survive'], alpha=0.7)
axes[1,0].set_title('Age Distribution: Survivors vs Non-Survivors', fontsize=14, fontweight='bold')
axes[1,0].set_xlabel('Age')
axes[1,0].legend()

# Plot 4: Survival by Embarkation Port
embark_survival = titanic.groupby('Embarked')['Survived'].size()
axes[1,1].bar(embark_survival.index, embark_survival.values * 100, color=['red', 'green', 'blue'])
axes[1,1].set_title('Survival Rate by Embarkation Port', fontsize=14, fontweight='bold')
axes[1,1].set_ylabel('Survival Rate (%)')

plt.tight_layout()
plt.show()
```

## 3.segregating X and y(features and target)

```python
## select target as survived
```

```python
x=titanic.drop(columns={'Survived'})
y=titanic['Survived']
```

## 4.Train and Test Split data


```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y)
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

#### i.encoding 

```python
## doing the onehotencoding for 'embarked' column,we use OHE bcz it is label category
### sex column is encoding by doing the OHE
```

```python
titanic['Embarked'].unique()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe= OneHotEncoder(sparse_output=False).set_output(transform='pandas')
ohe
```

#### ii.Scaling

```python
##### If number is category / label â†’ Donâ€™t Scale(pclass,sibsp,parch-->few unique values dont scale)
##### If number is continuous measurement â†’ Scale(age,fare)
```

```python
###### numeric columns-->Age,Fare
```

```python
fig,ax=plt.subplots(1,2,figsize=(20,4))
sns.boxplot(data=x_train,y='Age',ax=ax[0]) ## age contain outliers so we use RobustScaler
sns.boxplot(data=x_train,y='Fare',ax=ax[1])## fare cantains outliers so we use RobustScaler
```

```python
from sklearn.preprocessing import RobustScaler
```

```python
robust=RobustScaler()
robust
```

```python
## ColumnTransform -->at a time scaling and encoding is possible
```

```python
from sklearn.compose import ColumnTransformer
```

```python
ct=ColumnTransformer([('ohe',ohe,['Embarked','Sex']),('scale',robust,['Age','Fare'])],
                     verbose_feature_names_out=False,remainder='passthrough').set_output(transform='pandas')
ct
```

```python
x_train_transformed=ct.fit_transform(x_train)
x_train_transformed.head()
```

```python
#### before going to model building ,remove the text contain columns from x_train_transformed
#### bcz algo understand numbers only
```

```python
x_train_transformed.drop(columns={'Name','Ticket','Cabin'},inplace=True)
```

```python
x_train_transformed.head(2)
```
